```python


import pandas as pd

def parse_single_file(file_path):
    def parse_batch_file(file):
        with open(file, 'r') as f:
            lines = f.readlines()

        data = []
        for line in lines:
         fields = line.strip().split("\t")
         if len(fields) > 2 and "avesa.00003a" in fields[2] and "r1" in fields[2]:
                 transcript_id = fields[2].replace("'", "")
                 mercator_levels = fields[1].replace("'", "")
                 mercator_levels = mercator_levels.split(".")
                 data.append([transcript_id] + mercator_levels)

        if not data:  # Check if data is empty
            return pd.DataFrame(columns=["oat_transcript_id", "level1"])  # Return an empty DataFrame

        max_levels = max([len(row) - 1 for row in data])
        columns = ["oat_transcript_id"] + [f"level{i}" for i in range(1, max_levels + 1)]
        return pd.DataFrame(data, columns=columns)

    return parse_batch_file(file_path)

# Specify the file path you want to parse
single_file_path = "/home/sepidehjafarian/Desktop/oat.project/parsing results for protein function/oat_protein function.longest.mercator.results.txt"

# Parse a single file
parsed_data = parse_single_file(single_file_path)

# Modify transcript IDs
#parsed_data["oat_transcript_id"] = parsed_data["oat_transcript_id"].str.replace("h", "H")
#parsed_data["oat_transcript_id"] = parsed_data["oat_transcript_id"].str.replace("cHr", "chr")
parsed_data.head()

```
